h0: the mean beats per minute of a song in 2019 is 120
h1: the mean beats per minute of a song in 2019 is not equal to 120

```python
alpha=0.025
y = df['Beats.Per.Minute'].astype(int)
y.mean()    #population mean
```

```python
n=40   #sample size
for i in range(100):
    mu=y.sample(n=40).mean()    #sample mean
    print(mu)
```

```python
s=y.std()
s                  #standad deviation
```

```python
z=(y-mu)/(s/np.sqrt(n))
z.mean()
```

```python
z_critical = abs(st.norm.ppf(alpha))
z_critical
```

```python
n = 10
degrees_of_freedom = n-1
xbar = 120
alpha = 0.05
```

```python
m=y.sample(n=10).mean()
print(m)
stnd=y.std()
print(stnd)
```

```python
t = (xbar - m)/(stnd/ np.sqrt(n))
t
```

```python
t_critical = st.t.ppf(alpha/2, degrees_of_freedom)
t_critical
```

```python
if (abs(t) > abs(t_critical)): #Absolute value taken as the it's a two-tailed test and the original t_critical value might be negative
    print("Reject null hypothesis") 
else:
    print("Null hypothesis cannot be rejected")
```
