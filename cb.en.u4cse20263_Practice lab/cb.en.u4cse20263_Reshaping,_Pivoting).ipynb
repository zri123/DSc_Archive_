```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dayanandv/Data-Science/main/dataset/weather.csv')
df
```

# Reshaping

## Melt

```python
df1 = pd.melt(df, id_vars=['day'])
df1
```

Filter on city

```python
df1[df1['variable']=='chennai']
```

Changing variable and value

```python
melted = pd.melt(df, id_vars=['day'], var_name='city', value_name='temperature')
melted
```

```python
df_unmelted = melted.pivot(index='day', columns='city')
df_unmelted
```

# Pivot

## Basics

```python
df3 = pd.read_csv('https://raw.githubusercontent.com/dayanandv/Data-Science/main/dataset/weather_pivot.csv')
df3
```

```python
df3.pivot(index='city', columns='date')
```

```python
df3.pivot(index='city',columns='date',values="humidity")
```

## Table

```python
df3.pivot_table(index='city', columns='date', aggfunc='sum')
```

## Margins

```python
import numpy as np
df3.pivot_table(index='city', columns='date', margins=True, aggfunc=np.sum) # sum, diff
```

## Grouper

```python
df3
```

```python
df3['date']=pd.to_datetime(df3['date'])
df3.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```
