Q4

Outlier analysis for Transportation Expense

```python
q = ab_df["Transportation expense"].describe().loc[['25%', '75%']]
IQR = q[1] - q[0]
min_max = ab_df["Transportation expense"].describe().loc[['min', 'max']]
fil = (ab_df["Transportation expense"] < (1.5*IQR - min_max[0])) | (ab_df["Transportation expense"] > (min_max[1] + 1.5*IQR))
print(f'Outlier Count = {ab_df.loc[fil, "Transportation expense"].count()}')
```

Outlier Analysis for Height

```python
q1 = ab_df["Hit target"].describe().loc[['25%', '75%']]
IQR1 = q1[1] - q1[0]
min_max1 = ab_df["Hit target"].describe().loc[['min', 'max']]
fil1 = (ab_df["Hit target"] < (1.5*IQR1 - min_max1[0])) | (ab_df["Hit target"] > (min_max1[1] + 1.5*IQR1))
print(f'Outlier Count = {ab_df.loc[fil, "Hit target"].count()}')
```

Q14

```python
high_school = ab_df['Education'] == 1.0
grad = ab_df['Education'] == 2.0
post_grad = ab_df['Education'] == 3.0
avg_hs_age = ab_df.loc[high_school, "Age"].mean()
avg_grad_age = ab_df.loc[grad, "Age"].mean()
avg_pg_age = ab_df.loc[post_grad, "Age"].mean()
std_hs_age = ab_df.loc[high_school, "Age"].std()
std_grad_age = ab_df.loc[grad, "Age"].std()
std_pg_age = ab_df.loc[post_grad, "Age"].std()
print(f'Average age of highschoolers = {avg_hs_age}')
print(f'Average age of graduates = {avg_grad_age}')
print(f'Average age of postgraduates = {avg_pg_age}')
print(f'Deviation age of highschoolers = {std_hs_age}')
print(f'Deviation age of graduates = {std_grad_age}')
print(f'Deviation age of postgraduates = {std_pg_age}')
```
